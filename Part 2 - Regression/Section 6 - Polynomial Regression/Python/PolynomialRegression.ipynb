```python
import numpy as mp
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:,-1].values
```

```python
# Test Linear Regression model first
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X,y)
```

```python
#Visualize Linear Regression
plt.scatter(X, y, color='red')
plt.plot(X, regressor.predict(X),color='blue')
plt.title('LinearRegression Model')
plt.xlabel('Level of Employment')
plt.ylabel('Salary')
```

```python
#Polynomial regression
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=4)
X_poly = pf.fit_transform(X)
LinearRegression2 = LinearRegression()
LinearRegression2.fit(X_poly,y)
```

```python
# visualise polynomial regression
plt.scatter(X, y, color='red')
plt.plot(X, LinearRegression2.predict(X_poly),color='blue')
plt.title('LinearRegression Model')
plt.xlabel('Level of Employment')
plt.ylabel('Salary')
```

```python
#prediction from linear regression
print(regressor.predict([[6.5]]))
#prediction from polynomial regression
print(LinearRegression2.predict(pf.transform([[6.5]])))
```
